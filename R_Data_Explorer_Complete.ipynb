```r
# Create a list (in R, this type of list is called a 'vector') of character column names that need conversion to numeric.
# 'c' is a function in R that combines its arguments to form a vector.

columns_to_convert <- c("agricultural_land", "cpi_change", "forested_area",
                        "gasoline_price", "gross_primary_education_enrollment",
                        "gross_tertiary_education_enrollment", "minimum_wage", "out_of_pocket_health_expenditure",
                        "population_labor_force_participation", "tax_revenue",
                        "total_tax_rate", "unemployment_rate")

# Convert the specified character columns to numeric by:
# 1. Removing symbols like "$" and "%".
# 2. Replacing commas with dots (to make them valid numeric values).
# The Mutate and Across functions from dplyr are used to apply these changes to each specified column.
# The modified dataset, with the selected columns transformed to numeric format, is stored in the new variable clean_world.

clean_world <- world_data %>%
  mutate(across(all_of(columns_to_convert),
                ~ as.numeric(gsub(",", ".", gsub("[\\$%]", "", .)))))
```

Normal Distribution (Bell-shaped)

- Data is symmetrically distributed around the mean.
Implication: This is an ideal distribution for many statistical methods. Data that follows this shape might not need transformations before modeling.
Right-skewed (Positively skewed)

-  The tail on the right-hand side is longer than the left. Most of the data is concentrated on the left.
Implication: The mean is typically greater than the median in this distribution. Log-transformations can sometimes help in normalizing the data.
Left-skewed (Negatively skewed)

-  The tail on the left-hand side is longer than the right. Most of the data is concentrated on the right.
Implication: The mean is typically less than the median. Square or cube transformations might help in normalizing the data.
Uniform Distribution

- Data is evenly distributed across the range of the data.
Implication: Every value in the range is equally likely to occur.
Bimodal Distribution

- Data has two peaks.
Implication: This might suggest that the data is a mix of two different groups.

```r
# View quantile-quantile plot of all continuous variables, escluding Latitude and Longitude
#A Quantile-Quantile (QQ) plot is a graphical tool used to compare the quantiles of two data distributions.
# Typically, one of the distributions is theoretical (like the normal distribution), and the other is the empirical distribution of the dataset.


clean_world_sub <- cleaned_world %>% #esclude lat and lng from QQ (not good candidates for QQ analysis)
  select(-latitude, -longitude) %>%
  na.omit()


plot_qq(na.omit(clean_world_sub),
  title = "Quantile-Quantile Plots- IQR",
  ggtheme = theme_minimal(),
  geom_qq_args = list(color = "dodgerblue", size = 2),
  geom_qq_line_args = list(color = "red", linetype = "dashed"),  # Adjust QQ line appearance
  theme_config = list(axis.title = element_text(size = 10, face = "bold"),
                      axis.text = element_text(size = 8)),
  nrow = 4L,
  ncol = 4L,
  parallel = FALSE
)

```

If the data points lie approximately on the dashed red line, it means the empirical distribution of the data matches the theoretical normal distribution.

Deviations from the line indicate departures from normality. This can be useful to understand the nature of the data distribution and consider potential transformations if needed.

1. **Notebook**: Imagine a digital notebook, similar to a diary, where instead of just writing text, you can also run small pieces of code, display graphs, and show results. It's an interactive environment that allows for both explanation (like writing notes) and execution (like doing math problems) in one place.

2. **Kernel**: Think of the kernel as the brain behind the notebook. It's the engine that understands and runs the code you write. If the notebook is our diary, then the kernel is the pen that brings our thoughts (or code) to life.

3. **R Language**: R is a popular language primarily used for statistical computing and graphics. Picture it as a specialized language that excels in understanding and talking about data, much like how a musician might be adept in the language of music.

4. **Packages**: In the R language, a package is like a toolbox filled with various tools (functions) designed for specific tasks. Just as a carpenter might have a toolbox for woodworking, a data analyst using R has packages tailored for data visualization, analysis, and more.

5. **Libraries**: The term "library" can sometimes be used interchangeably with "package." But to be more precise, when we "load a library" in R, it's like opening that toolbox (package) and making all its tools (functions) available for use in our current project.

6. **Dataset**: Picture a dataset as a digital spreadsheet or a big table filled with rows and columns of information. Just as a librarian might have a collection of books with different information, analysts and researchers have datasets containing various details, like sales numbers, customer feedback, or weather records. It's the foundational set of information that they work on to derive insights, make predictions, or understand patterns.

7. **Functions**: Think of a function like a mini-machine or a kitchen gadget. You give it certain ingredients (inputs), and it processes them to give you a finished product (output). For example, a juicer takes fruit and produces juice. In the digital world, a function might take numbers and calculate their sum for you.

8. **Variables**: A variable is like a labeled jar or container where you can store something for use later. For instance, you might have a jar labeled "Cookies" where you keep your cookies. In the coding world, if you have data or a value you want to use later, you store it in a variable with a specific name, so you can easily find and use it when needed.

9. **Object**: In R, an object is a data structure having some attributes and methods which act on its attributes. Everything in R is considered an object, whether it's data or a function.

10. **Data Frame**: A data frame is a table or a two-dimensional array-like structure in R. It contains data with rows and columns. Each column can be of different types (e.g., numbers, text), and a data frame is the most commonly used object for data analysis in R.

11. **Plot**: A visual representation or diagram that showcases data in a graphical format. Plots are designed to provide insights, highlight patterns, or reveal relationships within the data. Examples include bar graphs, line charts, and scatter plots. Think of plots as pictures that tell a story about numbers or data points.

12. **Vector** :(in R) a vector is a basic data structure that contains elements of the same type. The function 'c' is used in R to create a vector. For instance, c(1,2,3,4) creates a numeric vector with elements 1, 2, 3, and 4. Vectors can be used to represent lists of items like names of columns, values, or even characters.
